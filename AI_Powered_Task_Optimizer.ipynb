```python
import pandas as pd
from datetime import datetime, timedelta
import random
```

```python
# Sample task data generation
titles = ["Write report", "Study for test", "Finish project", "Team meeting", "Reply to emails", "Code review"]
priorities = ["High", "Medium", "Low"]
data = []
```

```python
for i in range(20):
    task = {
        "Title": random.choice(titles),
        "Description": "Some task details",
        "Deadline": (datetime.now() + timedelta(days=random.randint(1, 10))).strftime('%Y-%m-%d'),
        "Estimated_Duration_Hrs": random.randint(1, 8),
        "User_Priority": random.choice(priorities)
    }
    data.append(task)
```

```python
df = pd.DataFrame(data)
df.to_csv("task_data.csv", index=False)
print("Generated synthetic task data.")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from datetime import datetime
```

```python
# Load data
data = pd.read_csv("task_data.csv")
```

```python
# Feature engineering
data["Deadline"] = pd.to_datetime(data["Deadline"])
data["Days_Left"] = (data["Deadline"] - datetime.now()).dt.days
data["Duration_Category"] = pd.cut(data["Estimated_Duration_Hrs"], bins=[0, 2, 5, 10], labels=[0, 1, 2])
```

```python
# Encode labels
le = LabelEncoder()
data["Priority_Label"] = le.fit_transform(data["User_Priority"])
```

```python
# Define features and target
X = data[["Days_Left", "Estimated_Duration_Hrs"]]
y = data["Priority_Label"]
```

```python
# Train/test split â€” now should work!
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# Predict
data["Predicted_Priority_Label"] = model.predict(X)
data["Predicted_Priority"] = le.inverse_transform(data["Predicted_Priority_Label"])
```
